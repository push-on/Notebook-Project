# Pandas & Numpy

```python
import pandas as pd
import numpy as np
```

# Loading Dataset

```python
data = pd.read_csv("../assets/heart-disease.csv", na_filter=False)
print(data.head())
print(data.dtypes.nunique())
```

```python
# One Hot Encoding
pd.get_dummies(data)
```

# Handle missing values

```python
# Handle missing values
data = data.fillna(data.mean())
data
```

# Data Cleaning

```python
# Data Cleaning
data = data.drop_duplicates()
data
```

# Data Transformation

```python
# Data Transformation
data = pd.get_dummies(data, columns=["age"])
data
```

# Normalization


```python
# Normalization
data = (data - data.mean()) / data.std()
data
```

# Data Splitting


```python
# Data Splitting
train_data = data.iloc[:int(data.shape[0] * 0.8), :]
test_data = data.iloc[int(data.shape[0] * 0.8):, :]

data
```
