## Open

```python
from PIL import Image
import os

image_path = "./me.jpg"

if not os.path.exists(image_path):
    print(" Image was not found")

image = Image.open(image_path)

# image.show()
display(image)
```

## Save 

```python
output_image_path = "./me_2.jpg"

image.save(output_image_path)
output_image = Image.open(output_image_path)

if os.path.exists(output_image_path):
    print(" Image created successfully!")
else:
    print(" Image creation failed.")

output_image.close()

```

```python
if os.path.exists(output_image_path):
    os.remove(output_image_path)
    print(" Removed Previously Created Image ")
else:
    print(" Image does not exist.")
```

##  Find a  rgb value of a pixel

```python
x = 200  # x-coordinate of the pixel
y = 200  # y-coordinate of the pixel

rgb_value = image.getpixel((x, y))
print(rgb_value)

canvas = Image.new("RGB", (100, 100), "white")
pixels = canvas.load()
for i in range(100):
    for j in range(100):
        pixels[i, j] = rgb_value

display(canvas)
```

## Change the color of a portion of an image

```python
x = 50  
y = 50 
width = 50  
height = 70  
modified_image = image.copy()

for i in range(x, x + width):
    for j in range(y, y + height):
        modified_image.putpixel((i, j), (107, 179, 105))

x = 55
y = 52
width = 40
height = 40

for i in range(x, x + width):
    for j in range(y, y + height):
        modified_image.putpixel((i, j), (154, 232, 149))

display(modified_image)
```

## Convert a BGR image to grayscale

```python
grayscale_image = modified_image.convert("L")
display(grayscale_image)
```
