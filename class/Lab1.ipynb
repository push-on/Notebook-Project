## ðŸ“¤ Reading & showing image in Notebook

```python
from PIL import Image

image_path = "./me.jpg"
image = Image.open(image_path)
# image.show()
display(image)
```

## ðŸ”Ž finding img format pil lib

```python
image_type = image.format
image_type
```

## ðŸ¤ height, width & channels of img

```python
image_height = image.height
image_width = image.width
image_channels = len(image.getbands())
print("Image height:", image_height)
print("Image width:", image_width)
print("Image channels:", image_channels)
```

## ðŸ”¨ Resize the image

```python
new_width = 300
new_height = 300
resized_image = image.resize((new_width, new_height))
image_height = resized_image.height
image_width = resized_image.width
print("Resized Image height:", image_height)
print("Resized Image width:", image_width)
display(resized_image)
```

## ðŸ³ï¸â€ðŸŒˆ RGB image to ðŸ”³ grayscale

```python
grayscale_image = resized_image.convert("L")
display(grayscale_image)
```
